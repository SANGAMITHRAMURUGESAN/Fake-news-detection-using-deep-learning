```python
# Take 1000 random samples from the dataset
samples = test_dataset.shuffle(seed=42).select(range(12))  # change to 1000 later to scale up
dataset = samples.map(lambda sample: {"text": format_instruction(sample)})
print(dataset)
dataset = dataset.remove_columns(['input'])
print(dataset)
print(dataset['text'][0])
print(dataset['output'][0])
```

```python
import gc


# Define batch size
batch_size = 4  # Adjust


# Set a padding token if not already set
#if tokenizer.pad_token is None:
#    tokenizer.pad_token = tokenizer.eos_token

# Define a maximum sequence length for your model
max_length = 4096#512

predictions = []
actual_labels = []




```

```python
predictions = []
actual_labels = []
raw_outputs = []



for i in range(0, len(dataset)):
    gc.collect()  # Garbage collection

    text = dataset[i]['text']
    label = dataset[i]['output']

    # Tokenize texts in batch with padding and truncation
    input = tokenizer(text, truncation=True, return_tensors="pt")

    input_ids = input.input_ids

    # Move input data to GPU
    #input_ids = input_ids.to(device)

    # Generate outputs
    #with torch.no_grad():
    outputs = model.generate(input_ids, max_new_tokens=2)  # Reduced max_new_tokens


    decoded_output = tokenizer.batch_decode(outputs.detach().cpu().numpy(), skip_special_tokens=True)
    print("Raw model output:", decoded_output)  # Debugging print
    raw_outputs.append(decoded_output)

    prediction = decoded_output[0][len(text):]
    print(f"Generated Prediction:\n{prediction}")
    prediction = int(prediction.strip())
    predictions.append(prediction)

    actual_labels.append(label)

```

```python
print(predictions)
predictions = [int(x.strip()) for x in predictions]
print(predictions)
```

```python
from sklearn.metrics import classification_report

# Calculate accuracy and generate the classification report
accuracy = sum(torch.tensor(predictions) == torch.tensor(actual_labels)) / len(actual_labels)
report = classification_report(actual_labels, predictions, target_names=['Class 0', 'Class 1'])

print(f'Accuracy: {accuracy}')
print("\nClassification Report:\n", report)
```
